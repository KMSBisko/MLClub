```python
import numpy as np
def manual_cosine_similarity(list1, list2):
    # Create a combined set of words
    combined_set = list(set(list1) | set(list2))

    # Create frequency vectors for both lists
    vector1 = np.array([list1.count(word) for word in combined_set])
    vector2 = np.array([list2.count(word) for word in combined_set])

    # Calculate cosine similarity
    cosine_similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
    return cosine_similarity
```

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline

device = "cuda:0" if torch.cuda.is_available() else "cpu"
#device = "cpu" 
#torch_dtype = torch.float32
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

print(device)
print(torch_dtype)
# Load the model and processor
model_name = "openai/whisper-large-v3-turbo"
model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_name, torch_dtype=torch_dtype
)  # Load the model onto the GPU
model.to(device)
processor = AutoProcessor.from_pretrained(model_name)

# Use the GPU by setting the device to 'cuda'
pipe = pipeline("automatic-speech-recognition",
                model=model, tokenizer=processor.tokenizer, feature_extractor=processor.feature_extractor, device="cuda")
```

```python
# Provide the path to the audio file
sample = "Untitled notebook.wav" #put whatever file you want to transcribe here, demo with Travis Scott's 'Impossible'

# Run the pipeline using the GPU
result = pipe(sample, return_timestamps=True)
# Print the result
file1 = open("actual_lyrics.txt", "r") #actual text to check model accuracygit
model_output = result["text"]
actual_text = file1.read()
sentences = model_output.split('.')
for sentence in sentences:
    if sentence.strip():
        print(sentence.strip())
print("Model accuracy stats using cosine similarity:")
print(manual_cosine_similarity(actual_text, model_output))
```
